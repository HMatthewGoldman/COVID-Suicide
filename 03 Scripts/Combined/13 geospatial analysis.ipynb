## Plotting a Choropleth with Geoplot

### Plotting covid cases by state

```python
# Setup a folium map at a high-level zoom
map_age_covid = folium.Map(location = [100, 0], zoom_start = 1.5)

# Choropleth maps bind Pandas Data Frames and json geometries.This allows us to quickly visualize data combinations
folium.Choropleth(
    geo_data = usa_geo, 
    data = state_age_covid,
    columns = ['state', 'covid_by_age'],
    key_on = 'feature.properties.name', # this part is very important - check your json file to see where the KEY is located
    fill_color = 'YlOrBr', fill_opacity=0.6, line_opacity=0.1,
    legend_name = "covid_by_age").add_to(map_age_covid)
folium.LayerControl().add_to(map_age_covid)
map_age_covid.save('plot_data.html')

map_age_covid
```

```python
# Setup a folium map at a high-level zoom
map_sex_covid = folium.Map(location = [100, 0], zoom_start = 1.5)

# Choropleth maps bind Pandas Data Frames and json geometries.This allows us to quickly visualize data combinations
folium.Choropleth(
    geo_data = usa_geo, 
    data = state_sex_covid,
    columns = ['state', 'covid_by_sex'],
    key_on = 'feature.properties.name', # this part is very important - check your json file to see where the KEY is located
    fill_color = 'RdYlBu', fill_opacity=0.6, line_opacity=0.1,
    legend_name = "covid_by_sex").add_to(map_sex_covid)
folium.LayerControl().add_to(map_sex_covid)
map_sex_covid.save('plot_data.html')

map_sex_covid
```

```python
# Setup a folium map at a high-level zoom
map_ethnicity_covid = folium.Map(location = [100, 0], zoom_start = 1.5)

# Choropleth maps bind Pandas Data Frames and json geometries.This allows us to quickly visualize data combinations
folium.Choropleth(
    geo_data = usa_geo, 
    data = state_ethnicity_covid,
    columns = ['state', 'covid_by_ethnicity'],
    key_on = 'feature.properties.name', # this part is very important - check your json file to see where the KEY is located
    fill_color = 'PiYG', fill_opacity=0.6, line_opacity=0.1,
    legend_name = "covid_by_ethnicity").add_to(map_ethnicity_covid)
folium.LayerControl().add_to(map_ethnicity_covid)
map_ethnicity_covid.save('plot_data.html')

map_ethnicity_covid
```
