```python
# create gender dataframe
df3=df.drop(df.index[df['Ethnicity_1'].isin(['American Indian/Alaska Native', 'Black', 'Hispanic', 'White',
       'Asian/Pacific Islander'])])
df3.reset_index(inplace=True)
df3
```

```python
df3.drop(columns=['index'],inplace=True)
```

```python
df3.rename(columns={'Ethnicity_1':'sex'},inplace=True)
```

```python
df3=df3.iloc[:, [0,1,3,2]]
```

```python
df3.head()
```

```python
df3.sort_values(by=['year','state','sex'])
```

```python
# create dataframe groups to match other dataframes
dfgroup=df3.groupby(['year','state','sex'])[['suicide_deaths']].agg('sum')
```

```python
# round suicide_deaths to 2 decimal placed
dfgroup['suicide_deaths']=dfgroup['suicide_deaths'].round(2)
```

```python
# check grouping dataframe
dfgroup.head(50)
```

```python
#Use sweetviz to generate report
my_report = sv.analyze(df3)
my_report.show_html('Gender Suicides 2019-2021.html')
```
